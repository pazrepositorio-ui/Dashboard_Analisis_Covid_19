##### <u>Paso 8: Limpieza datos, registro nulos y correcci√≥n tipo de dato</u>

```python
# Completar los valores faltantes con el valor anterior (forward fill) por pa√≠s
for col in df_Latinoamerica_mayor_2021_locationkey.columns:
    if df_Latinoamerica_mayor_2021_locationkey[col].isna().sum() > 0 and col != 'country_name':
        df_Latinoamerica_mayor_2021_locationkey[col] = (
            df_Latinoamerica_mayor_2021_locationkey
            .groupby('country_name')[col]
            .transform(lambda x: x.fillna(method='ffill')))
```

```python
# verifico que ninguna columna tenga nulos
df_Latinoamerica_mayor_2021_locationkey.isnull().sum()
```

```python
# Rellenar cumulative_vaccione_doses_administrered con ffill + bfill por pa√≠s
df_Latinoamerica_mayor_2021_locationkey["cumulative_vaccine_doses_administered"] = (
    df_Latinoamerica_mayor_2021_locationkey
    .groupby("country_name")["cumulative_vaccine_doses_administered"]
    .transform(lambda x: x.fillna(method='ffill').fillna(method='bfill')))

# Rellenar new_recovered con ffill + bfill por pa√≠s
df_Latinoamerica_mayor_2021_locationkey["new_recovered"] = (
    df_Latinoamerica_mayor_2021_locationkey
    .groupby("country_name")["new_recovered"]
    .transform(lambda x: x.fillna(method='ffill').fillna(method='bfill')))

# Rellenar cumulative_recovered con ffill + bfill por pa√≠s
df_Latinoamerica_mayor_2021_locationkey["cumulative_recovered"] = (
    df_Latinoamerica_mayor_2021_locationkey
    .groupby("country_name")["cumulative_recovered"]
    .transform(lambda x: x.fillna(method='ffill').fillna(method='bfill')))
```

```python
# verifico que ninguna columna tenga nulos
df_Latinoamerica_mayor_2021_locationkey.isnull().sum()
```

```python
# Se rellenan con 0 aquellas columnas que aun tienen nulos - Todos los datos son nulos para un pais
df_Latinoamerica_mayor_2021_locationkey["new_recovered"] = df_Latinoamerica_mayor_2021_locationkey["new_recovered"].fillna(0)
df_Latinoamerica_mayor_2021_locationkey["cumulative_recovered"] = df_Latinoamerica_mayor_2021_locationkey["cumulative_recovered"].fillna(0)
```

```python
# verifico que ninguna columna tenga nulos
df_Latinoamerica_mayor_2021_locationkey.isnull().sum()
```

##### <u>Paso 12: <u>Funci√≥n mediana, varianza y rango</u>

```python
# Defino una funci√≥n que tiene como argumento una columna, que luego puedo reutilizar
def obtener_estadisticas(columna):
    # Devuelve la mediana, varianza y rango de una serie de datos num√©ricos
    mediana = columna.median()
    varianza = columna.var()
    rango = columna.max() - columna.min()
    return mediana, varianza, rango
```

```python
# luego se usar√≠a asi
columna_vacunas_administradas = df_Latinoamerica_mayor_2021_locationkey['cumulative_vaccine_doses_administered']
mediana, varianza, rango = obtener_estadisticas(columna_vacunas_administradas)

print(f'Mediana: {mediana}')
print(f'Varianza: {varianza}')
print(f'Rango: {rango}')
```

##### <u> EXTRA CREDIT: Funciones de orden superior - Map, Filter, Apply y Reduce</u>

### MAP()

```python
# map() aplica una funci√≥n a cada elemento de una secuencia y devuelve el resultado.
# Columna de casos confirmados y que se convierte en ‚Äúalto‚Äù o ‚Äúbajo‚Äù
nivel_contagio = map(lambda x: "alto" if x > 100000 else "bajo", df_Latinoamerica_mayor_2021_locationkey["new_confirmed"])
df_Latinoamerica_mayor_2021_locationkey["nivel_contagio"] = list(nivel_contagio) 
```

```python
# Visualizamos resultado
print(df_Latinoamerica_mayor_2021_locationkey[["country_name", "new_confirmed", "nivel_contagio"]].head(10))
```

### FILTER()

```python
# nombres de pa√≠s donde los casos nuevos confirmados son mayores a 1.000.000
paises_altos_niveles = list(filter(lambda pais: df_Latinoamerica_mayor_2021_locationkey[df_Latinoamerica_mayor_2021_locationkey["country_name"] == pais]["new_confirmed"].sum() > 1000000,
                           df_Latinoamerica_mayor_2021_locationkey["country_name"].unique()))
```

```python
# Ver qu√© pa√≠ses cumplieron la condici√≥n
print(paises_altos_niveles)
```

### REDUCE()

```python
# Reducir una lista a un valor √∫nico - suma total de casos confirmados para todos los registros
from functools import reduce
total_contagios = reduce(lambda a, b: a + b, df_Latinoamerica_mayor_2021_locationkey["new_confirmed"])

print(total_contagios)
```

### APPLY()

```python
# Crear columna categ√≥rica para casos nuevos (Podr√≠a haber usado MAP())
df_Latinoamerica_mayor_2021_locationkey["nivel_casos_nuevos"] = df_Latinoamerica_mayor_2021_locationkey["new_confirmed"].apply(
    lambda x: "bajo" if x < 5000 else ("moderado" if x < 10000 else "alto")
)

# Visualizaci√≥n
print(df_Latinoamerica_mayor_2021_locationkey[["new_confirmed", "nivel_casos_nuevos"]].head(10))
```

<h5 style="color: red;">Lineplot: Evoluci√≥n de dosis administradas - Muertes - Casos confirmados y recuperaciones por mes de cada pa√≠s</h5>

```python

# Copiar el dataframe original
df = df_Latinoamerica_mayor_2021_locationkey.copy()

# Asegurar columna fecha
df["date"] = pd.to_datetime(df["date"])

# Ordenar por pa√≠s y fecha
df = df.sort_values(["country_name", "date"])

# Calcular nuevas dosis entre d√≠as consecutivos por pais (es para desarmnar el acumulado). Reemplaza con 0 los primeros
df["new_vaccine_doses_administered"] = (
    df.groupby("country_name")["cumulative_vaccine_doses_administered"]
    .diff()
    .fillna(0)
)

# Crear columna a√±o-mes
df["year_month"] = df["date"].dt.to_period("M").astype(str)

# Agrupar por pa√≠s y mes
df_mensual = (
    df.groupby(["country_name", "year_month"])[
        ["new_confirmed", "new_deceased", "new_recovered", "new_vaccine_doses_administered"]
    ]
    .sum()
    .reset_index()
)

# Lista de pa√≠ses a graficar
paises = ["Argentina", "Brazil", "Chile", "Peru", "Mexico", "Colombia"]

# Bucle por pa√≠s
for pais in paises:
    df_pais = df_mensual[df_mensual["country_name"] == pais]

    fig, ax1 = plt.subplots(figsize=(14,6))
    ax2 = ax1.twinx()

    # L√≠nea 1: Nuevos Confirmados
    ax1.plot(
        df_pais["year_month"],
        df_pais["new_confirmed"],
        label="Nuevos Confirmados",
        color="#023859",
        marker="o"
    )

    # L√≠nea 2: Nuevos Fallecidos
    ax1.plot(
        df_pais["year_month"],
        df_pais["new_deceased"],
        label="Nuevos Fallecidos",
        color="#F23E2E",
        marker="o"
    )

    # L√≠nea 3: Nuevos Recuperados
    ax1.plot(
        df_pais["year_month"],
        df_pais["new_recovered"],
        label="Nuevos Recuperados",
        color="#C5BEBE",
        marker="o"
    )

    # L√≠nea 4: Dosis Aplicadas (eje secundario)
    ax2.plot(
        df_pais["year_month"],
        df_pais["new_vaccine_doses_administered"],
        label="Dosis Aplicadas",
        color="#27CDF2",
        marker="o"
    )

    # Etiquetas de ejes
    ax1.set_xlabel("Mes")
    ax1.set_ylabel("Casos")
    ax2.set_ylabel("Dosis Aplicadas")

    # T√≠tulo
    plt.title(f"Evoluci√≥n Mensual de Indicadores en {pais}")

    # Rotaci√≥n de etiquetas X
    ax1.set_xticks(range(len(df_pais["year_month"])))
    ax1.set_xticklabels(df_pais["year_month"], rotation=45)

    # Leyenda combinada abajo centrada
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(
        lines1 + lines2,
        labels1 + labels2,
        loc="upper center",
        bbox_to_anchor=(0.5, -0.15),
        ncol=2,
        frameon=False )

    plt.tight_layout()
    plt.show()
```

En t√©rminos generales, Brasil y M√©xico fueron los pa√≠ses que presentaron los mayores vol√∫menes de nuevos casos confirmados a lo largo del per√≠odo analizado, mientras que Chile y Per√∫ registraron los niveles m√°s bajos de contagios. <h5 style="color: red;">Mapa de calor de m√©tricas por pa√≠s</h5>

```python
# Agrupar por pa√≠s y calcular las m√©tricas para cada columna seleccionada 
df_mapa = (
    df_Latinoamerica_mayor_2021_locationkey
    .groupby("country_name")
    .agg({"population": "max", "cumulative_confirmed": "max", "cumulative_deceased": "max", "gdp_per_capita_usd": "mean",
        "human_development_index": "mean", "smoking_prevalence": "mean", "diabetes_prevalence": "mean",
        "average_temperature_celsius": "mean"}).reset_index()
)

# Renombrar columnas para que se vean claros los nombres
df_mapa = df_mapa.rename(columns={
    "population": "Poblaci√≥n",
    "cumulative_confirmed": "Casos Confirmados",
    "cumulative_deceased": "Fallecidos",
    "gdp_per_capita_usd": "GDP per c√°pita",
    "human_development_index": "IDH",
    "smoking_prevalence": "Tabaquismo",
    "diabetes_prevalence": "Diabetes",
    "average_temperature_celsius": "Temperatura Media"})

# Normalizar las columnas (pone todo en la misma escala de 0 a 1, para que se puedan comparar visualmente o sumar entre s√≠.)
df_mapa_norm = df_mapa.copy()
metricas = df_mapa.columns[1:]  # Todas menos 'country_name'

for col in metricas:
    min_val = df_mapa[col].min()  # el valor mas bajo se convierte en 0
    max_val = df_mapa[col].max()  # el valor mas alto se convierte en 1
    df_mapa_norm[col] = (df_mapa[col] - min_val) / (max_val - min_val)  # los vallores intermedios van entre 0 y 1

# Transformar a formato largo para heatmap - mapa de filas=pa√≠s y columnas=m√©trica
df_heatmap = df_mapa_norm.set_index("country_name")

# Crear figura
plt.figure(figsize=(10,6))

sns.heatmap(
    df_heatmap,
    annot=True,
    fmt=".2f",
    cmap="Blues",
    cbar_kws={"label": "Valor Normalizado"},
    linewidths=0.5)

plt.title("Mapa de Calor de M√©tricas por Pa√≠s")
plt.xlabel("M√©trica")
plt.ylabel("Pa√≠s")

# Cambiar tama√±o de las etiquetas de eje X (M√©tricas) y Y (Pa√≠ses)
plt.xticks(fontsize=9)
plt.yticks(fontsize=9)

plt.tight_layout()
plt.show()
```

Brasil es el pa√≠s con valores m√°ximos en poblaci√≥n, casos confirmados y fallecidos (todas en 1), mientras que Chile se destaca por el m√°ximo IDH y GDP per c√°pita (ambos en 1).
```python
# Agrupar por pa√≠s
df_comp = (
    df_Latinoamerica_mayor_2021_locationkey
    .groupby("country_name")
    .agg({
        "diabetes_prevalence": "mean",
        "adult_male_mortality_rate": "mean",
        "adult_female_mortality_rate": "mean"
    })
    .reset_index()
)

# Crear columna de tasa de mortalidad promedio
df_comp["mortalidad_promedio"] = (
    (df_comp["adult_male_mortality_rate"] + df_comp["adult_female_mortality_rate"]) / 2
)

# Ordenar por mortalidad promedio
df_comp = df_comp.sort_values(by="mortalidad_promedio", ascending=False)

# Posiciones en eje X
x = np.arange(len(df_comp["country_name"]))
width = 0.35  # ancho de barra

# Crear gr√°fico
fig, ax = plt.subplots(figsize=(6,4))

bars1 = ax.bar(
    x - width/2,
    df_comp["diabetes_prevalence"],
    width,
    label="Prevalencia Diabetes",
    color="#27CDF2"
)

bars2 = ax.bar(
    x + width/2,
    df_comp["mortalidad_promedio"],
    width,
    label="Tasa Mortalidad Promedio",
    color="#023859"
)

# Etiquetas
ax.set_xlabel("Pa√≠s")
ax.set_ylabel("Porcentaje / Tasa")
ax.set_title("Comparaci√≥n de Prevalencia de Diabetes y Tasa de Mortalidad Adulta por Pa√≠s")
ax.set_xticks(x)
ax.set_xticklabels(df_comp["country_name"], rotation=45)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# Crear columna de mes
df_Latinoamerica_mayor_2021_locationkey["Mes"] = df_Latinoamerica_mayor_2021_locationkey["date"].dt.to_period("M")

# Agrupar por mes sumando casos y obteniendo la √∫ltima cantidad acumulada de vacunas de cada pa√≠s
agg = df_Latinoamerica_mayor_2021_locationkey.groupby(["Mes"]).agg({
    "new_confirmed": "sum",
    "cumulative_vaccine_doses_administered": "sum",
    "population": "sum"  # suma de poblaci√≥n total de todos los pa√≠ses
}).reset_index()

# Calcular cobertura de vacunaci√≥n sobre la poblaci√≥n total LATAM
agg["cobertura_vacunacion_pct"] = agg["cumulative_vaccine_doses_administered"] / agg["population"] * 100

# Crear el gr√°fico
fig, ax1 = plt.subplots(figsize=(8,6))

# Barra de casos nuevos
color_cases = custom_palette[0]
ax1.bar(
    agg["Mes"].astype(str),
    agg["new_confirmed"],
    color=color_cases,
    alpha=0.6,
    label="Casos Nuevos Confirmados"
)
ax1.set_xlabel("Mes")
ax1.set_ylabel("Nuevos Casos Confirmados", color=color_cases)
ax1.tick_params(axis="y", labelcolor=color_cases)
plt.xticks(rotation=45)

# L√≠nea de cobertura de vacunaci√≥n
ax2 = ax1.twinx()
color_vac = custom_palette[1]
ax2.plot(
    agg["Mes"].astype(str),
    agg["cobertura_vacunacion_pct"],
    marker='o',
    color=color_vac,
    linewidth=2,
    label="Cobertura de Vacunaci√≥n (%)"
)
ax2.set_ylabel("Cobertura de Vacunaci√≥n (%)", color=color_vac)
ax2.tick_params(axis="y", labelcolor=color_vac)

# T√≠tulo y leyenda
plt.title("Relaci√≥n entre Cobertura de Vacunaci√≥n y Nuevos Casos Confirmados en LATAM")
fig.tight_layout()
fig.legend(loc="upper left", bbox_to_anchor=(0.1,0.9))
plt.show()
```

La cobertura de vacunaci√≥n creci√≥ de forma sostenida mientras los nuevos casos mostraron picos marcados, sobre todo a inicios de 2022, y luego disminuyeron.

<h5 style="color: red;">Evoluci√≥n mensual de nuevas dosis aplicadas y muertes por Pa√≠s</h5>

```python
# Crear columna de mes
df_Latinoamerica_mayor_2021_locationkey["Mes"] = df_Latinoamerica_mayor_2021_locationkey["date"].dt.to_period("M")

# Agrupar por pa√≠s y mes, tomando el valor m√°ximo acumulado (√∫ltimo registro del mes)
agg_vacunas = df_Latinoamerica_mayor_2021_locationkey.groupby(["country_name", "Mes"]).agg({
    "cumulative_vaccine_doses_administered": "max"
}).reset_index()

# Ordenar por pa√≠s y mes
agg_vacunas = agg_vacunas.sort_values(by=["country_name", "Mes"])

# Calcular la diferencia respecto del mes anterior por pa√≠s
agg_vacunas["new_doses"] = agg_vacunas.groupby("country_name")["cumulative_vaccine_doses_administered"].diff().fillna(0)

# Verificar resultado
print(agg_vacunas.head(20))
```

```python
import matplotlib.pyplot as plt

# Crear el gr√°fico
plt.figure(figsize=(8,5))

# Lista de pa√≠ses en el dataset
paises = agg_vacunas["country_name"].unique()

# Colores por pa√≠s
colores_paises = {
    "Argentina": "#F23E2E",
    "Brazil": "#023859",
    "Mexico": "#594D46",
    "Chile": "#27CDF2",
    "Peru": "#FF8C00",
    "Colombia": "#C7A6A6"
}

# Graficar cada pa√≠s
for pais in paises:
    data = agg_vacunas[agg_vacunas["country_name"] == pais]
    plt.plot(
        data["Mes"].astype(str),
        data["new_doses"],
        marker='o',
        linestyle='-',
        label=pais,
        color=colores_paises.get(pais, None)
    )

plt.title("Evoluci√≥n Mensual de Dosis Nuevas Aplicadas por Pa√≠s")
plt.xlabel("Mes")
plt.ylabel("Dosis Nuevas Aplicadas")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

<h5 style="color: red;">HDI vs Incidencia Ajustada por Poblaci√≥n (cada 100.000 hab.)</h5>

```python
# Agrupar por pa√≠s: suma de casos y tomar HDI (es constante por pa√≠s)
agg_hdi = df_Latinoamerica_mayor_2021_locationkey.groupby("country_name").agg({
    "new_confirmed": "sum",
    "population": "mean",  # Es constante por pa√≠s
    "human_development_index": "mean"  # Es constante por pa√≠s
}).reset_index()

# Calcular incidencia por cada 100.000 habitantes
agg_hdi["incidencia_100k"] = agg_hdi["new_confirmed"] / agg_hdi["population"] * 100000

# Verificamos la tabla
print(agg_hdi)


```

```python
# Crear scatter plot
plt.figure(figsize=(10,4))
scatter = plt.scatter(
    agg_hdi["human_development_index"],
    agg_hdi["incidencia_100k"],
    s=100,
    color="#1f77b4",
    alpha=0.7
)

# Etiquetas de pa√≠s
for i, row in agg_hdi.iterrows():
    plt.text(
        row["human_development_index"] + 0.002,
        row["incidencia_100k"],
        row["country_name"],
        fontsize=9
    )

plt.title("Relaci√≥n entre √çndice de Desarrollo Humano e Incidencia Acumulada de COVID-19")
plt.xlabel("Human Development Index (HDI)")
plt.ylabel("Casos acumulados por cada 100.000 habitantes")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

Los pa√≠ses con mayor √çndice de Desarrollo Humano, como Chile y Argentina, registraron una incidencia acumulada m√°s alta de casos por cada 100.000 habitantes. En cambio, pa√≠ses con HDI m√°s bajo, como M√©xico, Colombia y Per√∫, presentaron valores menores. Esto sugiere que un mayor desarrollo humano puede estar asociado a una mayor capacidad de detecci√≥n y registro de casos, m√°s que necesariamente a mayor propagaci√≥n.
