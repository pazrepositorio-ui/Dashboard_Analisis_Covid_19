##### <u>Paso 8: Limpieza datos, registro nulos y corrección tipo de dato</u>

```python
# Completar los valores faltantes con el valor anterior (forward fill) por país
for col in df_Latinoamerica_mayor_2021_locationkey.columns:
    if df_Latinoamerica_mayor_2021_locationkey[col].isna().sum() > 0 and col != 'country_name':
        df_Latinoamerica_mayor_2021_locationkey[col] = (
            df_Latinoamerica_mayor_2021_locationkey
            .groupby('country_name')[col]
            .transform(lambda x: x.fillna(method='ffill')))
```

```python
# verifico que ninguna columna tenga nulos
df_Latinoamerica_mayor_2021_locationkey.isnull().sum()
```

```python
# Rellenar cumulative_vaccione_doses_administrered con ffill + bfill por país
df_Latinoamerica_mayor_2021_locationkey["cumulative_vaccine_doses_administered"] = (
    df_Latinoamerica_mayor_2021_locationkey
    .groupby("country_name")["cumulative_vaccine_doses_administered"]
    .transform(lambda x: x.fillna(method='ffill').fillna(method='bfill')))

# Rellenar new_recovered con ffill + bfill por país
df_Latinoamerica_mayor_2021_locationkey["new_recovered"] = (
    df_Latinoamerica_mayor_2021_locationkey
    .groupby("country_name")["new_recovered"]
    .transform(lambda x: x.fillna(method='ffill').fillna(method='bfill')))

# Rellenar cumulative_recovered con ffill + bfill por país
df_Latinoamerica_mayor_2021_locationkey["cumulative_recovered"] = (
    df_Latinoamerica_mayor_2021_locationkey
    .groupby("country_name")["cumulative_recovered"]
    .transform(lambda x: x.fillna(method='ffill').fillna(method='bfill')))
```

```python
# verifico que ninguna columna tenga nulos
df_Latinoamerica_mayor_2021_locationkey.isnull().sum()
```

```python
# Se rellenan con 0 aquellas columnas que aun tienen nulos - Todos los datos son nulos para un pais
df_Latinoamerica_mayor_2021_locationkey["new_recovered"] = df_Latinoamerica_mayor_2021_locationkey["new_recovered"].fillna(0)
df_Latinoamerica_mayor_2021_locationkey["cumulative_recovered"] = df_Latinoamerica_mayor_2021_locationkey["cumulative_recovered"].fillna(0)
```

```python
# verifico que ninguna columna tenga nulos
df_Latinoamerica_mayor_2021_locationkey.isnull().sum()
```

##### <u>Paso 12: <u>Función mediana, varianza y rango</u>

```python
# Defino una función que tiene como argumento una columna, que luego puedo reutilizar
def obtener_estadisticas(columna):
    # Devuelve la mediana, varianza y rango de una serie de datos numéricos
    mediana = columna.median()
    varianza = columna.var()
    rango = columna.max() - columna.min()
    return mediana, varianza, rango
```

```python
# luego se usaría asi
columna_vacunas_administradas = df_Latinoamerica_mayor_2021_locationkey['cumulative_vaccine_doses_administered']
mediana, varianza, rango = obtener_estadisticas(columna_vacunas_administradas)

print(f'Mediana: {mediana}')
print(f'Varianza: {varianza}')
print(f'Rango: {rango}')
```

##### <u> EXTRA CREDIT: Funciones de orden superior - Map, Filter, Apply y Reduce</u>

### MAP()

```python
# map() aplica una función a cada elemento de una secuencia y devuelve el resultado.
# Columna de casos confirmados y que se convierte en “alto” o “bajo”
nivel_contagio = map(lambda x: "alto" if x > 100000 else "bajo", df_Latinoamerica_mayor_2021_locationkey["new_confirmed"])
df_Latinoamerica_mayor_2021_locationkey["nivel_contagio"] = list(nivel_contagio) 
```

```python
# Visualizamos resultado
print(df_Latinoamerica_mayor_2021_locationkey[["country_name", "new_confirmed", "nivel_contagio"]].head(10))
```

### FILTER()

```python
# nombres de país donde los casos nuevos confirmados son mayores a 1.000.000
paises_altos_niveles = list(filter(lambda pais: df_Latinoamerica_mayor_2021_locationkey[df_Latinoamerica_mayor_2021_locationkey["country_name"] == pais]["new_confirmed"].sum() > 1000000,
                           df_Latinoamerica_mayor_2021_locationkey["country_name"].unique()))
```

```python
# Ver qué países cumplieron la condición
print(paises_altos_niveles)
```

### REDUCE()

```python
# Reducir una lista a un valor único - suma total de casos confirmados para todos los registros
from functools import reduce
total_contagios = reduce(lambda a, b: a + b, df_Latinoamerica_mayor_2021_locationkey["new_confirmed"])

print(total_contagios)
```

### APPLY()

```python
# Crear columna categórica para casos nuevos (Podría haber usado MAP())
df_Latinoamerica_mayor_2021_locationkey["nivel_casos_nuevos"] = df_Latinoamerica_mayor_2021_locationkey["new_confirmed"].apply(
    lambda x: "bajo" if x < 5000 else ("moderado" if x < 10000 else "alto")
)

# Visualización
print(df_Latinoamerica_mayor_2021_locationkey[["new_confirmed", "nivel_casos_nuevos"]].head(10))
```

<h5 style="color: red;">Heatmap: Matriz de correlación con variables clave</h5>

```python
# Definir variables seleccionadas
variables_clave = ['cumulative_confirmed', 'cumulative_deceased', 'population', 'average_temperature_celsius', 'human_development_index',
    'gdp_per_capita_usd', 'smoking_prevalence', 'diabetes_prevalence']

# Filtrar el DataFrame con esas variables
df_corr = df_Latinoamerica_mayor_2021_locationkey[variables_clave]

# Calcular la matriz de correlación
corr = df_corr.corr()

# Renombrar filas y columnas
nombres_nuevos = {
    'cumulative_confirmed': 'Confirmados Ac.',
    'cumulative_deceased': 'Muertes Ac.',
    'population': 'Población',
    'average_temperature_celsius': 'Temp. Promedio',
    'human_development_index': 'HDI',
    'gdp_per_capita_usd': 'GDP p/cápita',
    'smoking_prevalence': 'Tabaquismo',
    'diabetes_prevalence': 'Diabetes'}

corr = corr.rename(index=nombres_nuevos, columns=nombres_nuevos)

# Crear máscara para la parte superior
mask = np.triu(np.ones_like(corr, dtype=bool))

# Filtrar valores con correlación absoluta >0.5
filtered_corr = corr.where(corr.abs() > 0.5)

plt.figure(figsize=(8, 6))

# Crear el heatmap
ax = sns.heatmap(
    filtered_corr,
    mask=mask,
    annot=True,
    fmt=".2f",
    cmap="Blues",           
    vmin=-1,
    vmax=1,
    cbar_kws={"shrink": .8},
    linewidths=1,              
    linecolor="white",         
    annot_kws={"size":8})

# Cambiar tamaño y rotación de etiquetas
ax.set_xticklabels(ax.get_xticklabels(), fontsize=8, rotation=30, ha="right")
ax.set_yticklabels(ax.get_yticklabels(), fontsize=8, rotation=0)

# Título más grande y separado del gráfico
plt.title("Matriz de Correlación (solo valores >0.5)", fontsize=14, pad=15)

# Fondo gris claro alrededor
ax.set_facecolor('#f7f7f7')

# Ajustar layout
plt.tight_layout()

# Mostrar
plt.show()
```

<h5 style="color: red;">Lineplot: Evolución de dosis administradas - Muertes - Casos confirmados y recuperaciones por mes de cada país</h5>

```python

# Copiar el dataframe original
df = df_Latinoamerica_mayor_2021_locationkey.copy()

# Asegurar columna fecha
df["date"] = pd.to_datetime(df["date"])

# Ordenar por país y fecha
df = df.sort_values(["country_name", "date"])

# Calcular nuevas dosis entre días consecutivos por pais (es para desarmnar el acumulado). Reemplaza con 0 los primeros
df["new_vaccine_doses_administered"] = (
    df.groupby("country_name")["cumulative_vaccine_doses_administered"]
    .diff()
    .fillna(0)
)

# Crear columna año-mes
df["year_month"] = df["date"].dt.to_period("M").astype(str)

# Agrupar por país y mes
df_mensual = (
    df.groupby(["country_name", "year_month"])[
        ["new_confirmed", "new_deceased", "new_recovered", "new_vaccine_doses_administered"]
    ]
    .sum()
    .reset_index()
)

# Lista de países a graficar
paises = ["Argentina", "Brazil", "Chile", "Peru", "Mexico", "Colombia"]

# Bucle por país
for pais in paises:
    df_pais = df_mensual[df_mensual["country_name"] == pais]

    fig, ax1 = plt.subplots(figsize=(14,6))
    ax2 = ax1.twinx()

    # Línea 1: Nuevos Confirmados
    ax1.plot(
        df_pais["year_month"],
        df_pais["new_confirmed"],
        label="Nuevos Confirmados",
        color="#023859",
        marker="o"
    )

    # Línea 2: Nuevos Fallecidos
    ax1.plot(
        df_pais["year_month"],
        df_pais["new_deceased"],
        label="Nuevos Fallecidos",
        color="#F23E2E",
        marker="o"
    )

    # Línea 3: Nuevos Recuperados
    ax1.plot(
        df_pais["year_month"],
        df_pais["new_recovered"],
        label="Nuevos Recuperados",
        color="#C5BEBE",
        marker="o"
    )

    # Línea 4: Dosis Aplicadas (eje secundario)
    ax2.plot(
        df_pais["year_month"],
        df_pais["new_vaccine_doses_administered"],
        label="Dosis Aplicadas",
        color="#27CDF2",
        marker="o"
    )

    # Etiquetas de ejes
    ax1.set_xlabel("Mes")
    ax1.set_ylabel("Casos")
    ax2.set_ylabel("Dosis Aplicadas")

    # Título
    plt.title(f"Evolución Mensual de Indicadores en {pais}")

    # Rotación de etiquetas X
    ax1.set_xticks(range(len(df_pais["year_month"])))
    ax1.set_xticklabels(df_pais["year_month"], rotation=45)

    # Leyenda combinada abajo centrada
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(
        lines1 + lines2,
        labels1 + labels2,
        loc="upper center",
        bbox_to_anchor=(0.5, -0.15),
        ncol=2,
        frameon=False )

    plt.tight_layout()
    plt.show()
```

En términos generales, Brasil y México fueron los países que presentaron los mayores volúmenes de nuevos casos confirmados a lo largo del período analizado, mientras que Chile y Perú registraron los niveles más bajos de contagios. Esta misma tendencia se observa en la cantidad de nuevos recuperados, donde Brasil concentró la mayor cantidad de casos y Perú y Chile se ubicaron con cifras considerablemente menores. Respecto a los nuevos fallecidos, Brasil nuevamente se destacó por mantener los valores más altos de mortalidad mensual, en contraste con Chile y Argentina, que mostraron las líneas más bajas. Finalmente, al analizar las dosis de vacunas aplicadas, Brasil lideró con la mayor cantidad administrada en casi todos los meses, seguido de México, mientras que Chile y Perú fueron los países con menos dosis totales aplicadas en el período observado.

<h5 style="color: red;">Mapa de calor de métricas por país</h5>

```python
# Agrupar por país y calcular las métricas para cada columna seleccionada 
df_mapa = (
    df_Latinoamerica_mayor_2021_locationkey
    .groupby("country_name")
    .agg({"population": "max", "cumulative_confirmed": "max", "cumulative_deceased": "max", "gdp_per_capita_usd": "mean",
        "human_development_index": "mean", "smoking_prevalence": "mean", "diabetes_prevalence": "mean",
        "average_temperature_celsius": "mean"}).reset_index()
)

# Renombrar columnas para que se vean claros los nombres
df_mapa = df_mapa.rename(columns={
    "population": "Población",
    "cumulative_confirmed": "Casos Confirmados",
    "cumulative_deceased": "Fallecidos",
    "gdp_per_capita_usd": "GDP per cápita",
    "human_development_index": "IDH",
    "smoking_prevalence": "Tabaquismo",
    "diabetes_prevalence": "Diabetes",
    "average_temperature_celsius": "Temperatura Media"})

# Normalizar las columnas (pone todo en la misma escala de 0 a 1, para que se puedan comparar visualmente o sumar entre sí.)
df_mapa_norm = df_mapa.copy()
metricas = df_mapa.columns[1:]  # Todas menos 'country_name'

for col in metricas:
    min_val = df_mapa[col].min()  # el valor mas bajo se convierte en 0
    max_val = df_mapa[col].max()  # el valor mas alto se convierte en 1
    df_mapa_norm[col] = (df_mapa[col] - min_val) / (max_val - min_val)  # los vallores intermedios van entre 0 y 1

# Transformar a formato largo para heatmap - mapa de filas=país y columnas=métrica
df_heatmap = df_mapa_norm.set_index("country_name")

# Crear figura
plt.figure(figsize=(10,6))

sns.heatmap(
    df_heatmap,
    annot=True,
    fmt=".2f",
    cmap="Blues",
    cbar_kws={"label": "Valor Normalizado"},
    linewidths=0.5)

plt.title("Mapa de Calor de Métricas por País")
plt.xlabel("Métrica")
plt.ylabel("País")

# Cambiar tamaño de las etiquetas de eje X (Métricas) y Y (Países)
plt.xticks(fontsize=9)
plt.yticks(fontsize=9)

plt.tight_layout()
plt.show()
```

Brasil es el país con valores máximos en población, casos confirmados y fallecidos (todas en 1), mientras que Chile se destaca por el máximo IDH y GDP per cápita (ambos en 1). Argentina exhibe un valor alto en IDH (0,92), aunque en el resto de métricas se mantiene en rangos intermedios o bajos. México alcanza el valor máximo en prevalencia de diabetes (1) y un nivel medio en población. Perú resalta por una temperatura media elevada (0,85), pero en la mayoría de las métricas presenta valores reducidos. Colombia no muestra predominio claro en ninguna variable, manteniéndose en rangos medios o bajos.

<h5 style="color: red;">Relación entre cobertura de Vacunación y la Reducción de casos en LATAM</h5>

```python
# Crear columna de mes
df_Latinoamerica_mayor_2021_locationkey["Mes"] = df_Latinoamerica_mayor_2021_locationkey["date"].dt.to_period("M")

# Agrupar por mes sumando casos y obteniendo la última cantidad acumulada de vacunas de cada país
agg = df_Latinoamerica_mayor_2021_locationkey.groupby(["Mes"]).agg({
    "new_confirmed": "sum",
    "cumulative_vaccine_doses_administered": "sum",
    "population": "sum"  # suma de población total de todos los países
}).reset_index()

# Calcular cobertura de vacunación sobre la población total LATAM
agg["cobertura_vacunacion_pct"] = agg["cumulative_vaccine_doses_administered"] / agg["population"] * 100

# Crear el gráfico
fig, ax1 = plt.subplots(figsize=(8,6))

# Barra de casos nuevos
color_cases = custom_palette[0]
ax1.bar(
    agg["Mes"].astype(str),
    agg["new_confirmed"],
    color=color_cases,
    alpha=0.6,
    label="Casos Nuevos Confirmados"
)
ax1.set_xlabel("Mes")
ax1.set_ylabel("Nuevos Casos Confirmados", color=color_cases)
ax1.tick_params(axis="y", labelcolor=color_cases)
plt.xticks(rotation=45)

# Línea de cobertura de vacunación
ax2 = ax1.twinx()
color_vac = custom_palette[1]
ax2.plot(
    agg["Mes"].astype(str),
    agg["cobertura_vacunacion_pct"],
    marker='o',
    color=color_vac,
    linewidth=2,
    label="Cobertura de Vacunación (%)"
)
ax2.set_ylabel("Cobertura de Vacunación (%)", color=color_vac)
ax2.tick_params(axis="y", labelcolor=color_vac)

# Título y leyenda
plt.title("Relación entre Cobertura de Vacunación y Nuevos Casos Confirmados en LATAM")
fig.tight_layout()
fig.legend(loc="upper left", bbox_to_anchor=(0.1,0.9))
plt.show()
```

La cobertura de vacunación creció de forma sostenida mientras los nuevos casos mostraron picos marcados, sobre todo a inicios de 2022, y luego disminuyeron.

<h5 style="color: red;">Evolución mensual de nuevas dosis aplicadas y muertes por País</h5>

```python
# Crear columna de mes
df_Latinoamerica_mayor_2021_locationkey["Mes"] = df_Latinoamerica_mayor_2021_locationkey["date"].dt.to_period("M")

# Agrupar por país y mes, tomando el valor máximo acumulado (último registro del mes)
agg_vacunas = df_Latinoamerica_mayor_2021_locationkey.groupby(["country_name", "Mes"]).agg({
    "cumulative_vaccine_doses_administered": "max"
}).reset_index()

# Ordenar por país y mes
agg_vacunas = agg_vacunas.sort_values(by=["country_name", "Mes"])

# Calcular la diferencia respecto del mes anterior por país
agg_vacunas["new_doses"] = agg_vacunas.groupby("country_name")["cumulative_vaccine_doses_administered"].diff().fillna(0)

# Verificar resultado
print(agg_vacunas.head(20))
```

```python
import matplotlib.pyplot as plt

# Crear el gráfico
plt.figure(figsize=(8,5))

# Lista de países en el dataset
paises = agg_vacunas["country_name"].unique()

# Colores por país
colores_paises = {
    "Argentina": "#F23E2E",
    "Brazil": "#023859",
    "Mexico": "#594D46",
    "Chile": "#27CDF2",
    "Peru": "#FF8C00",
    "Colombia": "#C7A6A6"
}

# Graficar cada país
for pais in paises:
    data = agg_vacunas[agg_vacunas["country_name"] == pais]
    plt.plot(
        data["Mes"].astype(str),
        data["new_doses"],
        marker='o',
        linestyle='-',
        label=pais,
        color=colores_paises.get(pais, None)
    )

plt.title("Evolución Mensual de Dosis Nuevas Aplicadas por País")
plt.xlabel("Mes")
plt.ylabel("Dosis Nuevas Aplicadas")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

En el primer gráfico se observa que Brasil lideró de manera sostenida la aplicación de nuevas dosis durante la mayor parte del período, con un pico importante entre junio y septiembre de 2021. México también mostró un volumen alto, aunque más distribuido en el tiempo. Argentina tuvo un comportamiento particular con un pico muy marcado en mayo de 2022, que contrasta con su ritmo más estable en los meses anteriores. El resto de los países mantuvieron niveles más bajos y relativamente constantes de aplicación de dosis a lo largo del tiempo.

<h5 style="color: red;">HDI vs Incidencia Ajustada por Población (cada 100.000 hab.)</h5>

```python
# Agrupar por país: suma de casos y tomar HDI (es constante por país)
agg_hdi = df_Latinoamerica_mayor_2021_locationkey.groupby("country_name").agg({
    "new_confirmed": "sum",
    "population": "mean",  # Es constante por país
    "human_development_index": "mean"  # Es constante por país
}).reset_index()

# Calcular incidencia por cada 100.000 habitantes
agg_hdi["incidencia_100k"] = agg_hdi["new_confirmed"] / agg_hdi["population"] * 100000

# Verificamos la tabla
print(agg_hdi)


```

```python
# Crear scatter plot
plt.figure(figsize=(10,4))
scatter = plt.scatter(
    agg_hdi["human_development_index"],
    agg_hdi["incidencia_100k"],
    s=100,
    color="#1f77b4",
    alpha=0.7
)

# Etiquetas de país
for i, row in agg_hdi.iterrows():
    plt.text(
        row["human_development_index"] + 0.002,
        row["incidencia_100k"],
        row["country_name"],
        fontsize=9
    )

plt.title("Relación entre Índice de Desarrollo Humano e Incidencia Acumulada de COVID-19")
plt.xlabel("Human Development Index (HDI)")
plt.ylabel("Casos acumulados por cada 100.000 habitantes")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

Los países con mayor Índice de Desarrollo Humano, como Chile y Argentina, registraron una incidencia acumulada más alta de casos por cada 100.000 habitantes. En cambio, países con HDI más bajo, como México, Colombia y Perú, presentaron valores menores. Esto sugiere que un mayor desarrollo humano puede estar asociado a una mayor capacidad de detección y registro de casos, más que necesariamente a mayor propagación.
